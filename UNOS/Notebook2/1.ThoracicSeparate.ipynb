### User Function(s)

```python
def write_csv(df, filename):
    """
    write dataframe to disk
    """
    # intialize variable
    file_path = '../Data/' + filename
    
    # write to disk
    df.to_csv(file_path, index=False)
    
    return print(f"{len(df):,} records written to {file_path}.")
```

# Data Separate

```python
# extract adults only from main DataFrame
df_main_adult = df_main[df_main.AGE_GROUP == 'A'].copy()
```

```python
# display 
df_main_adult.head()
```

```python
# replace '.' with np.nan using map ( map operation across the entire DataFrame)
df_main_adult = df_main_adult.map(lambda x: np.nan if x == '.' else x)

# display random 10 rows
df_main_adult.sample(10)
```

```python
# check distinct values in the WL_ORG feature
df_main_adult.WL_ORG.value_counts(dropna=False)
```

```python
# remove all NaN from WL_ORG feature
df_main_adult = df_main_adult[~df_main_adult.WL_ORG.isnull()]
# recheck distinct values in the WL_ORG feature
df_main_adult.WL_ORG.value_counts(dropna=False)
```

```python
# write all the files
write_csv(df_main_adult, 'thoracic_all_main.csv')
# write HL
write_csv(df_main_adult[df_main_adult.WL_ORG == 'HL'], 'thoracic_HL_main.csv')
# write HR
write_csv(df_main_adult[df_main_adult.WL_ORG == 'HR'], 'thoracic_HR_main.csv')
# write LU
write_csv(df_main_adult[df_main_adult.WL_ORG == 'LU'], 'thoracic_LU_main.csv')
# write followup
write_csv(df_followup, 'thoracic_followup.csv')
```
